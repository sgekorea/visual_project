```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(df.index,df['출생아 수'])        # df에서 x축이 df.index(년도 나타내는거)  y축은 출생아수  파란색 선
plt.plot(df.index,df['합계 출산율'])   # 주황색 줄
```

```python
fig,ax1=plt.subplots(figsize=(15,10))
ax1.plot(df.index,df['출생아 수'],color='#ff812d')

ax2=ax1.twinx() # x축을 공유하는 쌍둥이 axi
ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
```

```python
fig,ax1=plt.subplots(figsize=(13,5))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff812d')

ax2=ax1.twinx() # x축을 공유하는 쌍둥이 axi
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
```

```python
fig,ax1=plt.subplots(figsize=(13,5))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff812d')
for idx,val in enumerate(df['출생아 수']):
    ax1.text(idx,val+12,val,ha='center')
ax2=ax1.twinx() # x축을 공유하는 쌍둥이 axi
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
```

```python
fig,ax1=plt.subplots(figsize=(13,5))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff812d')
for idx,val in enumerate(df['출생아 수']):
    ax1.text(idx,val+12,val,ha='center')
ax2=ax1.twinx() # x축을 공유하는 쌍둥이 axi
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',markersize=15,lw=5,mec='w',mew=3)
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx,val+0.08,val,ha='center')
```

```python
fig,ax1=plt.subplots(figsize=(13,5))
fig.suptitle('출생아 수 및 합계출산율') # 그래프 제목 설정
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff812d')
for idx,val in enumerate(df['출생아 수']): # df[넣고 싶은값]  val->적을 글자 데이터
    ax1.text(idx,val+12,val,ha='center')   #val + 숫자 -> 막대그래프 위에 얼마나 띄어서 숫자 표시하는지  ha= center->중앙 정리
ax2=ax1.twinx() # x축을 공유하는 쌍둥이 axi
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',markersize=15,lw=5,mec='w',mew=3) # 마커 관련 lw 선 굷기 mec->원 외관 선 색 mew-> 원 외곽선 두께
for idx, val in enumerate(df['합계 출산율']): # 합계 출산율 선 위에 값 표시
    ax2.text(idx,val+0.08,val,ha='center')
```
